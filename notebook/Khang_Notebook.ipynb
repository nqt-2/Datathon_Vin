```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
#Q8: A

df_orders = pd.read_csv('../data/orders.csv')

cancelled_orders = df_orders[df_orders['order_status'] == 'cancelled']

payment_method_count = cancelled_orders['payment_method'].value_counts()

payment_method_count_descending = (payment_method_count).sort_values(ascending = False)

print("Payment method count for cancelled orders:")
print(payment_method_count_descending)
```

```python
#Q9: A

df_returns = pd.read_csv('../data/returns.csv')
df_products = pd.read_csv('../data/products.csv')
df_order_items = pd.read_csv('../data/order_items.csv', dtype={'promo_id_2': str})

order_items_size = pd.merge(
    df_order_items,
    df_products[['product_id', 'size']],
    on='product_id'
)

sales_count = order_items_size.groupby('size').size()

returns_size = pd.merge(
    df_returns, 
    df_products[['product_id', 'size']], 
    on='product_id')
returns_count = returns_size.groupby('size').size()

return_rate = (returns_count / sales_count).sort_values(ascending=False)

df_return_rates = pd.DataFrame({
    'total_sales': sales_count,
    'total_returns': returns_count,
    'return_rate': returns_count / sales_count
}).sort_values('return_rate', ascending=False)

df_return_rates
```
