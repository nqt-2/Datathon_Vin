```python
#Q6: A

path = '../data/customers.csv'
customers = pd.read_csv(path)

path = '../data/orders.csv'
orders = pd.read_csv(path)

df_merged = pd.merge(orders, customers[ ['customer_id', 'age_group'] ])

total_orders = df_merged.groupby('age_group').size()
total_people = customers.groupby('age_group').size()

avr_people = total_orders / total_people
print(avr_people.idxmax())
```

```python
#Q7: C

path = '../data/order_items.csv'
order_items = pd.read_csv(path)

path = '../data/orders.csv'
orders = pd.read_csv(path)

path = '../data/sales.csv'
sales = pd.read_csv(path)

path = '../data/geography.csv'
geography = pd.read_csv(path)

# test thử với 2012-07-05 -> kiểm tra Tổng quanity x unit_price tại ngày này có bằng Revenue không -> Tách theo 3 miền

new_data = pd.merge( orders[[ 'order_id', 'order_date', 'zip' ]], order_items[[ 'order_id', 'quantity', 'unit_price' ]] )
new_data['revenue'] = new_data['quantity'] * new_data['unit_price']

new_data_2 = pd.merge( new_data[[ 'zip', 'revenue' ]], geography[[ 'zip', 'region' ]])
arr = new_data_2.groupby('region').sum()
print(arr['revenue'].idxmax())

#714669
```
